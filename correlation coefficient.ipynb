```python
import math
n=int(input())
#print(n)
a1=[]
a2=[]
a3=[]

for i in range(n):
    t=input().strip().split('\t')
    a1.append(float(t[0]))
    a2.append(float(t[1]))
    a3.append(float(t[2]))
def func(x,y):
    xy=[]
    x2=[]
    y2=[]
    for i in range(len(x)):
        xy.append(x[i]*y[i])
        x2.append(x[i]*x[i])
        y2.append(y[i]*y[i])
    numerator=(len(x)*sum(xy))-(sum(x)*sum(y))
    denominator=(((len(x)*sum(x2)-(sum(x)**2))**(0.5))*((len(y)*sum(y2)-(sum(y)**2))**(0.5)))
    sol=numerator/denominator
    sol=round(sol,2)
    print(sol)
func(a1,a2)
func(a2,a3)
func(a1,a3)
```
